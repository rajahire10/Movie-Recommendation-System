**Importing ast module for converting string objects into list objects**

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies.dropna(inplace=True)
```

**Converting dict values of 'genres' column into list**

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

**Converting dict values of 'genres' column into list**

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fantasy"}]')
```

**Here we made helper function to get names of 3 cast actor of each movie**

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

**For crew column we are only taking directors name thats why we made helper function to fetch names of directors**

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

**made helper function to remove space between words**

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

**Here we applied our function to columns cast,crew,genres,keywords to remove spaces between words**

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

**As overview column is in string format so to convert it into list we applied lambda function**

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

**Here we added one column to our dataset i.e tags column and it consits of all data of overview,genres,keywords,cast and crew
columns.Now all features of all columns become one tag for each perticular movie**

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

**Here in this model we used vector space method so for that we need to import Countvectorizer model from sklearn library**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

**AS to calculate distance between two siliar vectors we used cosine_similarty angle 
i.e.smaller the angle ,larger similar contant in tags**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

**We made helper function to recommend movies by similar contents**

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```

**Here you can see that movies recommended by similar content of movie 'Gandhi' are:1)Gandhi, My Father
                                                                                   2)The Wind That Shakes the Barley
                                                                                   3)A Passage to India
                                                                                   4)Guiana 1838
                                                                                   5)Ramanujan**

```python
recommend('Avatar')
```
